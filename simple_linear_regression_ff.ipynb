```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
df['total_spent'] = df['TV']+df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.regplot(data=df,x='total_spent',y = 'sales')
```

```python
X = df['total_spent']
y = df['sales']
```

```python

#we need the y = mx+b equation 
#hence we solve and find the beta values and mse being y = b1x + b0
```

```python
np.polyfit(X,y,deg=1)
```

```python
#we found the beta vales to be 0.04868788, 4.24302822

pot_spent = np.linspace(0,500,200)
predicted_sale = 0.04868788*pot_spent + 4.24302822
```

```python
sns.scatterplot(data=df,x='total_spent',y = 'sales')
plt.plot(pot_spent,predicted_sale,color='red')
```
